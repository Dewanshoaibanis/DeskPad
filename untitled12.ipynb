```python
import random
import shutil
from pathlib import Path

data_path = data_dir/"food-101"/"images"


target= ["baklava","bruschetta","ceviche"]
data_split = ["train","test"]
amount = 0.2
lvl_split = {}
ro = Path("3items_20")
for p in data_split:
  split_file = data_dir/"food-101"/"meta"/ f"{p}.txt"
  with open(split_file, "r") as f:
    entries = [line.strip() for line in f.readlines() if line.split("/")[0] in target]

  n_amount = round(len(entries)*amount)
  sk= random.sample(entries,n_amount)

  img_path = [pathlib.Path(str(data_path/pic)+".jpg") for pic in sk]
  lvl_split[p] = img_path

  # for pic in sk:
  #   name,img_lvl = pic.split("/",1)
  #   src= main_dir/name/f"{img_lvl}.jpg"
  #   dst_dir = ro/p/name
  #   dst_dir.mkdir(exist_ok=True,parents=True)
  #   shutil.copy(src,dst_dir/f"{img_lvl}.jpg")
  # print(f"[{p:5s}] done copying into `{ro/p}`\n")


```

```python
lvl_split["train"][:13]
```

```python
target_dir = f"../data_dir/backlava_ceviche_bruschetta{str(int(amount*100))}_percent"
pathl = pathlib.Path(target_dir)
pathl.mkdir(exist_ok=True,parents=True)
```

```python
for dirt in lvl_split.keys():
  for path in lvl_split[str(dirt)]:
    dest_dir = pathl / dirt / path.parent.stem / path.name
    if not dest_dir.parent.is_dir():
       dest_dir.parent.mkdir(parents=True,exist_ok=True)
    print(f"[INFO] Copying {path} to {dest_dir}...")
    shutil.copy2(path,dest_dir)
```

```python
zipfile = data_dir/"my_first_creation"
shutil.make_archive(zipfile,"zip",pathl)
```
